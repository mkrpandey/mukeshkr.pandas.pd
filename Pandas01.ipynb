# #Series

```python
#Series #01
#Simple Series Function

import pandas as pd
x=[1,2,3,4,5,6]
y=pd.Series(x)
print(y)
print(type(y))
print(y[2])
```

```python
#Series (Modify) #02

import pandas as pd
x=[1,2,3,4,5,6]
y=pd.Series(x,index=['a','b','c','d','e','f'],dtype="float",name="python")
print(y)
print(type(y))
```

```python
#Series #03
#By dictionary function

import pandas as pd
dic={"name-":['python','c','java','c++','HTML'],"Pop-":[0,1,2,3,4,5],"Rank-":[1,4,3,2,5]}
x=pd.Series(dic)
print(x)
```

```python
#Series #04

s=pd.Series(12)
print(s)
print(type(s))
```

```python
#Series #05

s=pd.Series(12,index=[1,2,3,4,5])
print(s)
print(type(s))
```

```python
#Series #06

s=pd.Series(12,index=[1,2,3,4,5])
s1=pd.Series(12,index=[1,2,3])
print(s+s1)
```

# #Merge & Concat DataFrames

#Merge

```python
#Create 1st data frame with dic--// var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
#Create 1st data frame with dic--//var2=pd.DataFrame({"A":[11,22,33,44],"B":[55,66,77,88]})
#pd.merge(var1,var2)
#without parameter


import pandas as pd
var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,4],"C":[5,12,21,32]})

pd.merge(var1,var2)
```

```python
#Create 1st data frame with dic--// var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
#Create 1st data frame with dic--//var2=pd.DataFrame({"A":[11,22,33,44],"B":[55,66,77,88]})
#pd.merge(var1,var2,on="A") here "A" is common part, must be same data in both data frame
#with (on) parameter


import pandas as pd
var1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
var2=pd.DataFrame({"A":[1,2,3,4],"C":[5,12,21,32]})

pd.merge(var1,var2,on="A")
```

#Concat

```python
#Create 1st data series--// var1=pd.Series([1,2,3,4])
#Create 1st data series--//var2=pd.Series([5,6,7,8])
#pd.concat([var1,var2])
#without parameter


import pandas as pd
var1=pd.Series([1,2,3,4])
var2=pd.Series([5,6,7,8])

pd.concat([var1,var2])
```

# #GroupBy

```python
#Create a random data in/by DataFrame

import pandas as pd
var =pd.DataFrame({"name":["a","b","c","d","a","b","a","b","a","c","c"],
                   "S_1":[12,13,14,12,13,14,15,23,25,16,10],
                   "S_2":[23,24,25,26,27,28,29,30,25,34,35]})

var
```

```python
#Grouping

var_1=var.groupby("name")
var_1
```

```python
#Grouping by GroupBy 

for x,y in var_1:
    print(x)
    print(y)
    print()
```

```python
# get single person of data by get_group

var_1.get_group("a")
```

```python
#get min data 

var_1.min()
```

```python
#get max data

var_1.max()
```

```python
#get mean data

var_1.mean()
```

# #Melt Function & Pivot Table

#Melt()

```python
import pandas as pd
var=pd.DataFrame({"Days":[1,2,3,4,5,6],"Eng":[18,15,16,17,19,20],"Maths":[20,19,18,17,16,15]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var,id_vars=["Days"])
```

```python
pd.melt(var,id_vars=["Days"],var_name="Subject")
```

```python
pd.melt(var,id_vars=["Days"],var_name="Subject",value_name="Marks")
```

#Pivot()

```python
import pandas as pd
var=pd.DataFrame({"Days":[1,2,3,4,5,6],"St_Nm":["a","b","a","a","b","b"],"Eng":[18,15,16,17,19,20],"Maths":[20,19,18,17,16,15]})

var
```

```python
var.pivot(index="Days",columns="St_Nm")
```

```python
var.pivot(index="Days",columns="St_Nm",values="Eng")
```

```python
var.pivot(index="Days",columns="St_Nm",values="Maths")
```

```python
import pandas as pd
var=pd.DataFrame({"Days":[1,1,1,2,2,2],
                  "St_Nm":["a","b","a","a","b","b"],"Eng":[18,15,16,17,19,20],"Maths":[20,19,18,17,16,15]})

var
```

```python
var.pivot_table(index="St_Nm",columns="Days",
                aggfunc="mean")
```

```python
var.pivot_table(index="St_Nm",columns="Days",
                aggfunc="sum")
```

```python
var.pivot_table(index="St_Nm",columns="Days",
                aggfunc="sum",margins="True")
```
